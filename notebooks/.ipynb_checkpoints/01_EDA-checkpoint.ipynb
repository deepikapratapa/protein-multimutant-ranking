## Imports, Paths & Settings

````python
import os, re, warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 100)
warnings.filterwarnings("ignore")

# Resolve project root even if running from /notebooks
PROJECT_ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
DATA_RAW = PROJECT_ROOT / "data" / "raw"
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"

print("Project root:", PROJECT_ROOT)
print("Raw data dir:", DATA_RAW)
print("Processed dir:", DATA_PROCESSED)

DATA_PROCESSED.mkdir(parents=True, exist_ok=True)
````

## Quick Structure Check 

````python
# Lists expected top-level folders to verify repo structure
expected = {
    "data": (PROJECT_ROOT / "data").exists(),
    "data/raw": DATA_RAW.exists(),
    "data/processed": DATA_PROCESSED.exists(),
    "notebooks": (PROJECT_ROOT / "notebooks").exists(),
    "src": (PROJECT_ROOT / "src").exists(),
    "results": (PROJECT_ROOT / "results").exists(),
    "docs": (PROJECT_ROOT / "docs").exists(),
}
expected
````

## Load Dataset 

````python
# Detect dataset path relative to project root
path = DATA_RAW / "s2648.csv"
if not path.exists():
    raise FileNotFoundError(f"Dataset not found at: {path}\nPlace s2648.csv under data/raw/")

def try_read(p, sep):
    try:
        return pd.read_csv(p, sep=sep)
    except Exception:
        return None

# Try tab first, then comma, then auto sniff
df = try_read(path, sep="\t")
if df is None or (df.columns.size == 1 and df.shape[0] > 1):
    df = try_read(path, sep=",")
if df is None:
    df = pd.read_csv(path, sep=None, engine="python")

print("Loaded dataset:", path)
print("Shape:", df.shape)
df.head()
````

## Schema & Missingness

````python
print("Columns:", list(df.columns))
display(df.sample(min(5, len(df))))
display(df.info())
print("\nMissing values per column:")
display(df.isna().sum())
````

## Standardize Column Names

````python
# Map to canonical names used in the project
rename_map = {
    "CLID": "clid",
    "PDB": "pdb_id",
    "CHAIN": "chain",
    "MUT": "mut",
    "DDG": "ddg_exp_kcal",
    "pH": "pH",
    "T": "temp_C",
}
df = df.rename(columns=rename_map)
df.columns = [c.strip() for c in df.columns]

required = ["pdb_id","chain","mut","ddg_exp_kcal"]
missing = [c for c in required if c not in df.columns]
if missing:
    print("Missing expected columns:", missing)
else:
    print("Standardized columns OK.")
df.head()
````

## Parse Mutation Codes (`mut` → `wt`, `res_index`, `mut_aa`)

````python
def parse_mut(s):
    """
    Expects codes like C30S: wt=C, position=30, mut=S.
    Returns (wt:str, pos:int, mut:str) or (nan, nan, nan) if not matched.
    """
    m = re.fullmatch(r"([A-Z])(\d+)([A-Z])", str(s).strip())
    if m:
        wt, pos, mt = m.groups()
        return wt, int(pos), mt
    return np.nan, np.nan, np.nan

parsed = df["mut"].apply(parse_mut)
df["wt"] = parsed.apply(lambda x: x[0])
df["res_index"] = parsed.apply(lambda x: x[1])
df["mut_aa"] = parsed.apply(lambda x: x[2])

# Reorder for readability
front = ["clid","pdb_id","chain","wt","res_index","mut_aa","ddg_exp_kcal","mut","pH","temp_C"]
cols = [c for c in front if c in df.columns] + [c for c in df.columns if c not in front]
df = df[cols]
df.head()
````

## Sanity Checks

````python
# Parse success
parsed_ok = df["wt"].notna().mean()
print(f"Parsed mutation tokens for {parsed_ok*100:.1f}% of rows.")

# Basic label stats
print("\nΔΔG summary (kcal/mol):")
display(df["ddg_exp_kcal"].describe())

# Counts
if "clid" in df.columns:
    print("\nTop CLIDs by count:")
    display(df["clid"].value_counts().head(10))

print("\nTop PDBs by count:")
display(df["pdb_id"].value_counts().head(10))

# Duplicate check at mutation granularity
dup_cols = ["pdb_id","chain","wt","res_index","mut_aa"]
if all(c in df.columns for c in dup_cols):
    dups = df.duplicated(subset=dup_cols).sum()
    print(f"\nDuplicate mutation rows: {dups}")
````

## ΔΔG Distribution

````python
plt.figure(figsize=(8,5))
sns.histplot(df["ddg_exp_kcal"].dropna(), bins=40, kde=True)
plt.xlabel("ΔΔG (kcal/mol)")
plt.ylabel("Frequency")
plt.title("Distribution of Experimental ΔΔG")
plt.axvline(0, color='red', linestyle='--', label='Neutral Stability')
plt.legend()
plt.show()

# Stability counts
stabilizing = (df['ddg_exp_kcal'] < 0).sum()
destabilizing = (df['ddg_exp_kcal'] > 0).sum()
neutral = (df['ddg_exp_kcal'] == 0).sum()
print(f"Stabilizing (<0): {stabilizing}")
print(f"Destabilizing (>0): {destabilizing}")
print(f"Neutral (=0): {neutral}")
````

## CLID Frequencies (Histogram)

````python
if "clid" in df.columns:
    plt.figure(figsize=(12,5))
    sns.countplot(data=df, x="clid", order=df["clid"].value_counts().index)
    plt.xticks(rotation=90)
    plt.title("Frequency of CLID Entries")
    plt.xlabel("CLID (Protein ID)")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    print("Top 10 CLIDs by count:")
    display(df["clid"].value_counts().head(10))
else:
    print("'clid' column not found.")
````

## Mean ΔΔG per CLID (Protein Stability Trends)

````python
if "clid" in df.columns:
    mean_ddg = df.groupby("clid")["ddg_exp_kcal"].mean().sort_values(ascending=False).head(10)
    plt.figure(figsize=(10,4))
    sns.barplot(x=mean_ddg.index, y=mean_ddg.values)
    plt.title("Top 10 CLIDs by Mean ΔΔG (Higher = More Destabilizing)")
    plt.xlabel("CLID (Protein ID)")
    plt.ylabel("Mean ΔΔG (kcal/mol)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Top 10 CLIDs by Mean ΔΔG:")
    display(mean_ddg)
````

## ΔΔG vs Residue Position (Most Common CLID)

````python
if "res_index" in df.columns and "clid" in df.columns:
    top_clid = df["clid"].value_counts().idxmax()
    subset = df[df["clid"] == top_clid]
    plt.figure(figsize=(9,5))
    sns.scatterplot(data=subset, x="res_index", y="ddg_exp_kcal", alpha=0.7)
    plt.title(f"ΔΔG vs Residue Position — {top_clid}")
    plt.xlabel("Residue Index")
    plt.ylabel("ΔΔG (kcal/mol)")
    plt.axhline(0, color='red', linestyle='--')
    plt.show()
else:
    print("Columns missing: 'res_index' or 'clid'.")
````

##  pH and Temperature Effects 

````python
if "pH" in df.columns:
    plt.figure(figsize=(6,4))
    sns.scatterplot(data=df, x="pH", y="ddg_exp_kcal", alpha=0.7)
    plt.title("ΔΔG vs pH")
    plt.xlabel("pH")
    plt.ylabel("ΔΔG (kcal/mol)")
    plt.show()

if "temp_C" in df.columns:
    plt.figure(figsize=(6,4))
    sns.scatterplot(data=df, x="temp_C", y="ddg_exp_kcal", alpha=0.7)
    plt.title("ΔΔG vs Temperature (°C)")
    plt.xlabel("Temperature (°C)")
    plt.ylabel("ΔΔG (kcal/mol)")
    plt.show()
````

## Type Fixes & Deduplication

````python
# Ensure types are clean
df["res_index"] = df["res_index"].astype("Int64")  # nullable int
df["wt"] = df["wt"].astype("string")
df["mut_aa"] = df["mut_aa"].astype("string")
df["chain"] = df["chain"].astype("string")
df["pdb_id"] = df["pdb_id"].astype("string")
df["clid"] = df.get("clid", pd.Series([pd.NA]*len(df))).astype("string")

# Deduplicate at mutation granularity by averaging ΔΔG (and pH/T)
key = ["pdb_id","chain","wt","res_index","mut_aa"]
agg = {"ddg_exp_kcal":"mean"}
if "pH" in df.columns:     agg["pH"] = "mean"
if "temp_C" in df.columns: agg["temp_C"] = "mean"
if "clid" in df.columns:   agg["clid"] = "first"

clean = (df.groupby(key, dropna=False, as_index=False)
           .agg(agg)
           .rename(columns={"mut_aa":"mut"}))

print("Original rows:", len(df), "→ After dedup:", len(clean))
clean.head()
````

## Save Canonical Clean File

````python
cols_order = ["clid","pdb_id","chain","wt","res_index","mut","ddg_exp_kcal"]
if "pH" in clean.columns: cols_order.append("pH")
if "temp_C" in clean.columns: cols_order.append("temp_C")
clean = clean[cols_order]

out_path = DATA_PROCESSED / "single_mut_clean.csv"
clean.to_csv(out_path, index=False)
print("Saved:", out_path)
````

## Single-Protein Subset (Most Abundant CLID)

````python
if "clid" in df.columns:
    top_clid = df["clid"].value_counts().idxmax()
    subset = clean[clean["clid"] == top_clid].copy()
    sp_path = DATA_PROCESSED / f"single_mut_{top_clid}.csv"
    subset.to_csv(sp_path, index=False)
    print("Saved single-protein subset:", sp_path, " (n=", len(subset), ")")
else:
    print("No 'clid' column to subset by.")
````

## Train/Test Split with Stratification (by protein + label sign)

````python
# Train/Test Split with robust label_sign and stratification (no 'string' dtype)

from sklearn.model_selection import train_test_split

tmp = clean.copy()

# treat tiny values as zero to avoid float noise
eps = 1e-9
tmp["label_sign"] = np.where(tmp["ddg_exp_kcal"] >  eps, "pos",
                      np.where(tmp["ddg_exp_kcal"] < -eps, "neg", "zero")).astype(str)

# Build stratification key: protein + sign (as plain str)
clid_str = tmp["clid"].fillna("NA").astype(str) if "clid" in tmp.columns else pd.Series(["NA"]*len(tmp))
strat_col = clid_str + "_" + tmp["label_sign"]

# Collapse very-rare strata (count < 2) to avoid errors
counts = strat_col.value_counts()
rare = counts[counts < 2].index
strat_col_collapsed = strat_col.where(~strat_col.isin(rare), other="OTHER")

train_idx, test_idx = train_test_split(
    tmp.index, test_size=0.2, random_state=42, stratify=strat_col_collapsed
)

train_df = tmp.loc[train_idx].drop(columns=["label_sign"])
test_df  = tmp.loc[test_idx ].drop(columns=["label_sign"])

train_df.to_csv(DATA_PROCESSED / "train_split.csv", index=False)
test_df.to_csv(DATA_PROCESSED / "test_split.csv", index=False)

print("Train/Test sizes:", len(train_df), len(test_df))
print("Strata kept (not OTHER):", (strat_col_collapsed != "OTHER").mean())
print("Train label balance:")
print(pd.cut(train_df["ddg_exp_kcal"], bins=[-np.inf, -eps, eps, np.inf], labels=["neg","zero","pos"]).value_counts())
````

## QA Plots on Cleaned Data
